```python
# ── Update these paths to match your local system ──
AUDIO_DIR    = r"C:\Users\kanis\Downloads\IIMSTC\2nd Mini Project\FSC22\Audio Wise V1.0-20220916T202003Z-001\Audio Wise V1.0"
METADATA_CSV = r"C:\Users\kanis\Downloads\IIMSTC\2nd Mini Project\FSC22\Metadata-20220916T202011Z-001\Metadata\Metadata V1.0 FSC22.csv"
OUTPUT_DIR   = r"C:\Users\kanis\Downloads\IIMSTC\2nd Mini Project\data\processed"
CSV_OUTPUT   = r"C:\Users\kanis\Downloads\IIMSTC\2nd Mini Project\data\metadata.csv"

# Audio configuration — FSC22 native specs
SAMPLE_RATE  = 44100   # FSC22 native sample rate
DURATION     = 5.0     # FSC22 native clip duration in seconds
TOP_DB       = 30      # Silence trimming threshold

print(f"Audio Directory  : {AUDIO_DIR}")
print(f"Metadata CSV     : {METADATA_CSV}")
print(f"Output Directory : {OUTPUT_DIR}")
print(f"Sample Rate      : {SAMPLE_RATE} Hz")
print(f"Clip Duration    : {DURATION} seconds")
```

```python
meta = pd.read_csv(METADATA_CSV)

print(f"Total files in metadata : {len(meta)}")
print(f"Columns                 : {meta.columns.tolist()}")
print(f"\nFirst 5 rows:")
print(meta.head())
print(f"\nAll 27 unique classes:")
print(meta['Class Name'].unique())
```

```python
print("Step 1: Checking for corrupted files...")
valid_files = []

for _, row in tqdm(meta.iterrows(), total=len(meta)):
    file_path = os.path.join(AUDIO_DIR, row["Dataset File Name"])
    try:
        audio, sr = librosa.load(file_path, sr=None)
        valid_files.append(row["Dataset File Name"])
    except:
        print(f"  [CORRUPT] {row['Dataset File Name']}")

print(f"\nValid files  : {len(valid_files)} / {len(meta)}")
print(f"Corrupt files: {len(meta) - len(valid_files)}")

# Filter metadata to valid files only
meta = meta[meta["Dataset File Name"].isin(valid_files)].reset_index(drop=True)
print(f"\nMetadata filtered to {len(meta)} valid files")
print("Corruption check complete — passing to Kanish for preprocessing")
```
